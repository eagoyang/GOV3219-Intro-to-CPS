## Subsetting dataframe


### approach 1: subsetting using labels


```python
# Selection of one column
mydata1 = df['jobcat']
mydata1
```

```python
# Selection of multiple columns
mydata2 = df[['jobcat','salary']]
mydata2
```

```python
# Selection of multiples rows with conditions
mydata3 = df[df['jobcat']==3]
mydata3
```

```python
mydata4 = df[df['salary'] > 50000]
mydata4
```

### approach 2: subsetting using location indexes


```python
# Selection of a single row using loc()
mydata5 = df.loc[2]
mydata5
```

```python
# Selection of multiples rows using loc()
mydata6 = df.loc[[2,4,6]]
mydata6
```

```python
# Selection of a range of rows using loc()
mydata7 = df.loc[3:6]
mydata7
```

```python
# Selection of rows and columns using loc()
mydata8 = df.loc[3:6, ['jobcat','salary']]
mydata8
```

```python
#Selection of rows and columns using iloc() location index
mydata9 = df.iloc[[1,4,5],[4,5]] #This is to selects rows 1, 4, and 5, as well as columns 2 and 6
mydata9
```
